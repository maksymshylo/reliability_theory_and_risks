````python
def G(x):
    return (1 - exp(-(x**2)))
````

````python
def sigma_C(n,alpha):
    w = np.random.uniform(0,1,size=n)
    ksi = (1/alpha)*np.power(-np.log(w),1/4)
    q = 1 - G(ksi)
    Q = np.sum(q)/n
    sigma2 = (np.sum(q**2) - n*(Q**2))/(n-1)
    return (sigma2,Q)
````

````python
n_list  = np.arange(100,10**5,10**3)
sigma_list = [sigma_C(i,alpha=1)[0] for i in n_list]
plt.plot(n_list,sigma_list)
plt.show()
````
